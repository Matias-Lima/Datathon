```python
import pandas as pd
import matplotlib.pyplot as plt

# Lendo o arquivo CSV
df = pd.read_excel('C:/Users/limam/Documents/GitHub/Datathon_2023/Dados/BTC_News_150.xlsx')
Data = pd.DataFrame(df)
```

```python
Data['Data'] = pd.to_datetime(Data['Data'])
Data.set_index('Data', inplace=True)
```

```python
Data = Data.drop(['Stock', 'Headline'], axis=1)
```

```python
# Grouping the BTC news data by date and calculating the mean for sentiment scores
btc_news_grouped = Data.groupby('Data').mean()
# Displaying the resulting DataFrame
btc_news_grouped
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler

# Selecionando as colunas necessárias e removendo valores nulos
df_filtered = result[['daily_index', 'Sentiment Score']]

# Aplicando PCA com apenas uma dimensão
pca_1d = PCA(n_components=1)
pca_1d_result = pca_1d.fit_transform(df_filtered)

# Normalizando os resultados para a escala de 1 a 100
scaler = MinMaxScaler(feature_range=(1, 100))
normalized_pca_1d = scaler.fit_transform(pca_1d_result)

# Convertendo o resultado em um dataframe
df_filtered['PCA_Normalized'] = normalized_pca_1d
```
