```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 500)
```

```python
dummy = pd.read_excel('C:/Users/limam/Documents/GitHub/Datathon_2023/Code/Python/result_PCA_indicador_dia.xlsx')
dummy.head(50)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(dummy['PCA_Normalized'], label='PCA_Index')
plt.plot(dummy['daily_index'], label='daily_index')
plt.title('Comparison of PCA_Index and Daily Index')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig = px.line(dummy, x=dummy.index, y='PCA_Normalized', markers=False)
fig.add_scatter(x=dummy.index, y=dummy['daily_index'], mode='lines', name='daily_index')
fig.update_layout(title='Combined Time Evolution', title_x=0.5, 
                  xaxis_title='Index', yaxis_title='Values', 
                  width=1500, height=600, legend=dict(x=0, y=1))
fig.update_xaxes(tickangle=70, tickformat='%Y-%m')
fig.show()
```
